### 3、图片ndarray

```python
from PIL import Image
```

```python
img = Image.open('1.jpg')
img
```

```python
dog_data = np.array(img)
# 图片数据是ndarray
# 彩色图片是三维的：高度，宽度，像素（表示不同颜色），红绿蓝三原色
dog_data
```

```python
dog_data.shape
```

```python
dog_data.dtype
```

```python
dog2 = dog_data[:,:,::-1] # 调整颜色
dog2 
```

```python
Image.fromarray(dog2) 
```

```python
Image.fromarray(dog_data[::2,:]) # 调整大小
```

```python
# 红绿蓝0,1,2
# 绿红蓝1,0,2
Image.fromarray(dog_data[:,:,[1,0,2]])
```

```python
dog3= dog_data[:,:,0]  # 颜色设置黑白
Image.fromarray(dog3)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.imshow(dog_data)
```

```python
plt.imshow(dog_data[::-15,::-15])  # 马赛克
```

### 5、级联

```python
plt.imshow(dog_data)
```

```python
dog9 = dog_data[:,:150]
plt.imshow(dog9)
```

```python
dog7 = dog_data[:,150:300]
plt.imshow(dog7)
```

```python
dog8 = dog_data[:,300:]
plt.imshow(dog8)
```

```python
print(dog9.shape,dog8.shape,dog7.shape)
```

```python
p =np.concatenate([dog8,dog7,dog9],axis=1)  # axis设置按哪条轴拼接（其他轴的shape必须相同）
plt.imshow(p)
```

```python
nd1 =  np.random.randint(0,150,size = (4,5))
nd2 =  np.random.randint(0,150,size = (2,5))
nd3 =  np.random.randint(0,150,size = (4,8))

display(nd1,nd2,nd3)
```

```python
np.concatenate([nd1,nd3],axis= 1)
```

```python
# 水平级联 -- 列数增加
nd4 = np.hstack((nd1,nd3))
nd4
```

```python
nd5 = np.vstack((nd1,nd2))
nd5
```

### 分割（级联逆操作）

```python
np.split(nd4,indices_or_sections=4) # 均分四份
```

```python
np.split(nd4,indices_or_sections=[1,3]) # 第一行切一刀，第三行切一刀
```

```python
np.vsplit(nd4,indices_or_sections=2) # 行切割
```

```python
np.hsplit(nd4,indices_or_sections=[3,8])# 列切割
```

## 四、ndarray聚合操作

```python
nd2
```

```python
nd2.sum() # 求和
```

```python
nd2[0].prod() # 累乘
```

```python
nd2.mean()
```

```python
nd2.mean(axis=0)
```

```python
nd2.argmax()
```

```python
nd2.argmax(axis=1)
```

```python
np.argwhere(nd2 > 90) # 返回nd2>90的数据的位置下标
```

```python
np.argwhere((nd2 > 90) & (nd2 < 120)) # 多条件逻辑连接符
```

```python
nd3 = nd2.ravel() # 展开成一维
```

```python
np.median(nd3) # 中位数
```

```python
np.percentile(nd3,q=1)  # 分位数，q参数指定百分数
```

```python
nd2.any() # nd2中元素全为False 返回False，否则True
```

```python
nd2.all() # nd2中元素全为True 返回True，否则False
```
