```python
# 删除空元素
df.dropna()
```

```python
# 删除列
df.drop(labels=['En'],axis = 1)  
```

```python
# 定位删除空元素
cond = df.isnull().any(axis = 1)
index = df[cond].index
df2 = df.drop(index)
df2
```

```python
 # 获取大于60分的数据
df3 = df2[df2>60]
df3.dropna()
```

```python
 # 获取大于60分的数据
cond = (df2 < 60).any(axis =1)
index = df2[cond].index
df2.drop(index)
```

```python
dfm = df2.mean(axis  = 1 )
dfm
```

```python
# 小于60或大于110数据获取
df3 = dfm[(dfm > 110) | (dfm < 60)]
df3.dropna()
```
