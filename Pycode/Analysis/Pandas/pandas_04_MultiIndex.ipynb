## 数据聚合【重点】

```python
df = DataFrame({'item' : np.random.randint(0,10,size=100),
                'sailer' : np.random.randint(0,10,size=100),
                'weight' : np.random.randint(30,200,size=100),
                'price' : np.random.randint(1,20,size = 100)})
df
```

```python
def convert_item(x):
    if x < 2:
        return '萝卜'
    elif x < 7:
        return '白菜'
    else:
        return '青椒'

df['item'] = df['item'].map(convert_item)
```

```python
df
```

```python
def convert_sailer(x):
    if x<5:
        return '张大妈'
    elif x<7:
        return '李大妈'
    else:
        return  '李大爷'
    
df['sailer'] = df['sailer'].map(convert_sailer)
```

```python
df
```

```python
def convert_weight(x):
    if x < 80:
        return 50
    elif x < 150:
        return 100
    else:
        return 150
    
df['weight'] = df['weight'].map(convert_weight)
```

```python
df
```

```python
# 数据分组
ret = df.groupby(['sailer'])['weight']
```

```python
# ret是一个分完组的对象
ret.sum()
```

```python
# apply函数后可调用其他函数，ret作为参数，返回值返回
ret.apply(np.sum)
```

```python
ret.max()
```

```python
ret.count()
```

```python
def analysis(x):
    return (x.size,x.sum(),np.round(x.mean(),decimals=2),x.min())
    
ret.apply(analysis)
```
