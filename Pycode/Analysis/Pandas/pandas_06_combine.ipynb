```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
```

## 一对一合并

```python
df1 = DataFrame({
    'id':[0,10,1024],
    'name':['bob','amy','rin'],
    'sex':['male','female','female']
})
df1
```

```python
df2 = DataFrame({
    'id':[0,11,1025],
    'salary':[10000,20000,30000],
    'age':[26,36,45]
})
df2
```

```python
df1.merge(df2)
```

## 多对一合并

```python
df2 = DataFrame({
    'id':[0,0,0],
    'salary':[10000,20000,30000],
    'age':[26,36,45]
})
df2
```

```python
df1.merge(df2)
```

## key的规范化

```python
df1 = DataFrame({
    'id':[0,10,1024],
    'name':['bob','amy','rin'],
    'sex':['male','female','female']
})


df2 = DataFrame({
    'id':[0,11,1025],
    'name':['mike','tim','frank'],
    'age':[26,36,45]
})

display(df1,df2)
```

```python
# on参数 指明根据某个索引进行融合
pd.merge(df1,df2,on='id')
```

```python
# 列冲突合并时，suffixes参数
pd.merge(df1,df2,on='id',suffixes=('_1','_2'))
```

```python
df2 = DataFrame({
    'ID':[0,120,110],
    'salary':[10000,20000,30000],
    'age':[26,36,45]
})
display(df1,df2)
```

```python
# 分别指明左右两边根据哪列索引进行合并
pd.merge(df1,df2,left_on='id',right_on='ID')
```

## 内合并与外合并

```python
pd.merge(df1,df2,left_on='id',right_on='ID',how='outer'  )
```

```python
# 左合并，保留左边数据
pd.merge(df1,df2,left_on='id',right_on='ID',how='left'  )
```

```python
dfr = DataFrame(np.random.randint(0,150,size=(8,3)),columns=['Python','Math','En'],index = list('ABCDEFGH'))
dfr
```

```python
df_s = DataFrame(dfr.sum(axis=1),columns=['sum'])
df_s
```

```python
# 根据左右两边的索引进行合并
pd.merge(dfr,df_s,left_index=True,right_index=True)
```
