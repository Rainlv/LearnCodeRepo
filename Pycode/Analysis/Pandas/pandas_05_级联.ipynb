```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
```

## np级联

```python
nd1 = np.random.randint(0,10,size= (4,5))
nd2 = np.random.randint(0,10,size= (7,5))
nd3 = np.random.randint(0,10,size= (4,8))
```

```python
# 哪条轴形状不同 按哪条轴级联
np.concatenate([nd1,nd2],axis = 0)
```

```python
np.concatenate([nd1,nd3],axis = 1)
```

## pd级联

```python
df1= DataFrame(np.random.randint(0,150,size=(5,3)),columns=['Python','Math','En'],index=list('ABCDE'))
df2= DataFrame(np.random.randint(0,150,size=(5,3)),columns=['Python','Math','En'],index=list('HIJKL'))

display(df1,df2)
```

```python
pd.concat([df1,df2])
```

```python
df3= DataFrame(np.random.randint(0,150,size=(5,3)),columns=['Chinese','Physic','Chem'],index=list('QWERT'))
df3
```

```python
df4= pd.concat([df1,df3],axis = 1)
df4.fillna(0)
```

```python
df5= DataFrame(np.random.randint(0,150,size=(5,3)),columns=['Python','Math','C++'],index=list('ABCGH'))
df5
```

```python
pd.concat([df1,df5],join='outer')
```

```python
# join参数默认是outer，设置成inner时，合并共同的索引项，抛弃非共同的
pd.concat([df1,df5],join='inner')
```

```python
# 以df1的索引进行合并
pd.concat([df1,df5],join_axes=[df1.columns])
```

```python
# ignore_index=True  忽略索引，重新创建（从0开始）
pd.concat([df1,df5],ignore_index=True)
```

```python
df6= DataFrame(np.random.randint(0,150,size=(5,3)),columns=['Python','Math','En'],index=list('ABCDE'))
df6
```

```python
pd.concat([df1,df6],keys=['期中','期末'])
```
