```python
import pandas as pd
from pandas import DataFrame
import numpy as np
```

```python
df = np.random.randint(0,150,size=(150,3))
df = DataFrame(df,columns=['Python','Math','En'])
df
```

```python
for i in range(15):
    index = np.random.randint(0,150,size=1)[0]
    column = np.random.randint(0,3,size=1)[0]
    df.iloc[index,column] = None
```

```python
df
```

```python
df.info()
```

```python
# 定位并取出含空数据的行列 
cond = df.isnull().any(axis = 1)
df[cond]  
```

```python
# 定位并取出非空数据的行列 
cond = df.notnull().all(axis = 1)
df[cond]
```

```python
# 删除空元素
df.dropna()
```

```python
# 删除列
df.drop(labels=['En'],axis = 1)  
```

```python
# 定位删除空元素
cond = df.isnull().any(axis = 1)
index = df[cond].index
df2 = df.drop(index)
df2
```

```python
 # 获取大于60分的数据
df3 = df2[df2>60]
df3.dropna()
```

```python
 # 获取大于60分的数据
cond = (df2 < 60).any(axis =1)
index = df2[cond].index
df2.drop(index)
```

```python
# 小于60或大于110数据获取
df3 = df2[(df2 > 110) | (df2 < 60)]
df3.dropna()
```
